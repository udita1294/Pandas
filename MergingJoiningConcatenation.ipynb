```python
import numpy as np
import pandas as pd
```

MERGING TWO DATAFRAMES

```python
employees = pd.DataFrame({
    'employee_id' : [1,2,3,4,5],
    'Name' : ['John','Anna','Peter','Linda','Bob'],
    'Department' : ['HR','IT','Finance','IT','HR']
})

salaries = pd.DataFrame({
    'employee_id' : [1,2,3,6,7],
    'Salary' : [600000,800000,650000,7000,90000],
    'Department' : [5000,10000,7000,8000,12000]
}) 
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees,salaries,on='employee_id',how='inner')
```

```python
pd.merge(employees,salaries,on='employee_id',how='outer')
```

```python
pd.merge(employees,salaries,on='employee_id',how='left')
```

```python
pd.merge(employees,salaries,on='employee_id',how='right')
```

CONCATENATION OF TWO DATAFRAMES

```python
df1 = pd.DataFrame({
    'A':['A0','A1','A2'],
    'B':['B0','B1','B3'],
    'C':['C0','C1','C2']
})

df2 = pd.DataFrame({
    'A':['A3','A4','A5'],
    'B':['B3','B4','B5'],
    'C':['C3','C4','C5']
})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df2,df1])
```

```python
pd.concat([df1,df2],axis=1)
```

 ** JOINING TWO DATAFRAMES**

```python
df1 = pd.DataFrame({
    'Name' : ['Alice','Bob','Charlie']
} , index=[1,2,3])

df2 = pd.DataFrame({
    'Score' : [85,90,75]
} , index=[2,3,4])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2)
```

```python
df1.join(df2,how='outer')
```

```python
df2.join(df1)
```

```python
df2.join(df1,how='outer')
```
